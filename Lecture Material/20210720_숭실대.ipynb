```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from tqdm import tqdm # for문의 진행상황

total = []

url = requests.get("https://search.naver.com/search.naver?where=nexearch&sm=top_hty&fbm=1&ie=utf8&query=%EC%97%B0%EA%B8%88%EB%B3%B5%EA%B6%8C")
html = BeautifulSoup(url.text)
current = int(html.find('a', attrs={'class':'_lottery-btn-current'}).find('em').text[:-1])

for n in tqdm(range(1, current + 1)):
  url = requests.get(f"https://search.naver.com/search.naver?where=nexearch&sm=top_hty&fbm=1&ie=utf8&query=%EC%97%B0%EA%B8%88%EB%B3%B5%EA%B6%8C+{n}%ED%9A%8C")
  html = BeautifulSoup(url.text)

  box = []
  numbers = html.find('ul', attrs={'class':'win_num'}).find_all('span')
  
  for i in numbers:
    box.append(int(i.text[0]))
  
  total.append(box)

df = pd.DataFrame(total, columns = ['조', '1번째번호', '2번째번호', '3번째번호', '4번째번호', '5번째번호', '6번째번호'])
df.to_excel('result.xlsx')
```

```python
url = requests.get("https://finance.naver.com/sise/sise_market_sum.nhn?sosok=0&page=1")
html = BeautifulSoup(url.text)
kospi_page = int(html.find('td', attrs={'class':'pgRR'}).find('a')['href'][-2:])

kospi_box = []

for n in tqdm(range(1, kospi_page + 1)):

    url = requests.get(f"https://finance.naver.com/sise/sise_market_sum.nhn?sosok=0&page={n}")
    html = BeautifulSoup(url.text)

    table = html.find('table', attrs={'class':'type_2'})
    table = pd.read_html(str(table))[0]
    table = table[table['종목명'].notnull()]
    del table['N']
    del table['토론실']
    table['종류'] = ['KOSPI'] * len(table)
    table['페이지'] = [n] * len(table)
    kospi_box.append(table)

kospi = pd.concat(kospi_box, ignore_index=True)

#-- 퀴즈 : 코스닥 크롤링하셔서 코스피랑 합쳐서 코스피+코스닥이 합쳐진 엑셀 만들기

url = requests.get("https://finance.naver.com/sise/sise_market_sum.nhn?sosok=1&page=1")
html = BeautifulSoup(url.text)
kosdaq_page = int(html.find('td', attrs={'class':'pgRR'}).find('a')['href'][-2:])

kosdaq_box = []

for n in tqdm(range(1, kosdaq_page + 1)):

    url = requests.get(f"https://finance.naver.com/sise/sise_market_sum.nhn?sosok=1&page={n}")
    html = BeautifulSoup(url.text)

    table = html.find('table', attrs={'class':'type_2'})
    table = pd.read_html(str(table))[0]
    table = table[table['종목명'].notnull()]
    del table['N']
    del table['토론실']
    table['종류'] = ['KOSDAQ'] * len(table)
    table['페이지'] = [n] * len(table)
    kosdaq_box.append(table)

kosdaq = pd.concat(kosdaq_box, ignore_index=True)

# 하나로 만들어진 코스피, 하나로 만들어진 코스닥 // 

stock = pd.concat([kospi, kosdaq], ignore_index=True)
stock.to_excel('stock.xlsx')
stock
```

```python
# 숭실대 -> 제목 10개 + 요약내용 10개 // 400페이지 4000개 뉴스

text_data = ''

for n in tqdm(range(1, 3992, 10)):

    url = requests.get(f"https://search.naver.com/search.naver?where=news&sm=tab_pge&query=%EC%88%AD%EC%8B%A4%EB%8C%80&sort=0&photo=0&field=0&pd=0&ds=&de=&cluster_rank=59&mynews=0&office_type=0&office_section_code=0&news_office_checked=&nso=so:r,p:all,a:all&start={n}")
    html = BeautifulSoup(url.text)

    news = html.find('ul', attrs={'class':'list_news'}).find_all('li', attrs={'class':'bx'})

    for i in news:
      title = i.find('a', attrs={'class':'news_tit'})['title']
      summary = i.find('div', attrs={'class':'dsc_wrap'}).text
      text_data += (title + ' ' + summary)
```

```python
import re # 정규표현식 : 원하는 패턴의 문자열을 효율적으로 찾는 방법

tel = '+1 (123) 456 7890'

# \d : 숫자
# \w : 문자
# \s : 공백
# ? : 있을수도있고 없을수도있다
# () : capture

re.findall('\+\d\s\(\d{3}\)\s?\d{3}\s?\d{4}', string=tel)
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt # 그래프, 그림 그리는 라이브러리

dic = {}

for i in re.findall('[가-힣]+', text_data):
  
  if i not in dic:
    dic[i] = 1
  else:
    dic[i] += 1

wc = WordCloud(
    font_path = 'BMDOHYEON_ttf.ttf',
    width = 800,
    height = 800,
    background_color = 'white'
).generate_from_frequencies(dic)

fig = plt.figure(figsize = (10, 10)) # 그림 그리기위한 도화지 까는것
plt.imshow(wc) # wc올리는 것
plt.show() # 보여주는것

# 정적 데이터
# 동적 데이터
```
